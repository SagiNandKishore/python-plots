```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

#print(city_df.head())
#print(ride_df.head())

# Combine the data into a single dataset
city_ride_df = ride_df.merge(city_df, how="left", on="city").reset_index().set_index("type")
city_ride_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x_urban = city_ride_df.loc["Urban"].groupby("city").count()["ride_id"]
y_urban = city_ride_df.loc["Urban"].groupby("city").mean()["fare"]
s_urban = city_ride_df.loc["Urban"].groupby(["city", "driver_count"]).count().reset_index().set_index("city")["driver_count"]

x_rural = city_ride_df.loc["Rural"].groupby("city").count()["ride_id"]
y_rural = city_ride_df.loc["Rural"].groupby("city").mean()["fare"]
s_rural = city_ride_df.loc["Rural"].groupby(["city", "driver_count"]).count().reset_index().set_index("city")["driver_count"]

x_suburban = city_ride_df.loc["Suburban"].groupby("city").count()["ride_id"]
y_suburban = city_ride_df.loc["Suburban"].groupby("city").mean()["fare"]
s_suburban = city_ride_df.loc["Suburban"].groupby(["city", "driver_count"]).count().reset_index().set_index("city")["driver_count"]

# Build the scatter plots for each city types

plt.scatter(x_urban, y_urban, marker = "o", facecolors= "tab:orange", s=s_urban, edgecolors = "black", label="Urban")
plt.scatter(x_rural, y_rural, marker = "o", facecolors= "tab:olive", s=s_rural , edgecolors = "black", label="Rural")
plt.scatter(x_suburban, y_suburban, marker = "o", facecolors= "tab:blue", s=s_suburban , edgecolors = "black", label="Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")
plt.grid(color='black', linestyle='-', linewidth=2, alpha = 0.2)

# Create a legend
plt.legend(loc=1)

# Incorporate a text label regarding circle size
plt.text(42,35,"Note:\nCircle size correlates with driver count per city")

# Save Figure
plt.savefig("../Images/PyberRideSharingData.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fare_df = city_ride_df.groupby("type").sum()

location_types=["Urban", "Rural", "Suburban"]
fares_list = [total_fare_df.loc[location]["fare"] for location in location_types]
colors = ["salmon", "gold", "lightskyblue"]
explode = [0.1,0,0]

# Build Pie Chart
fig = plt.pie(fares_list, 
              explode = explode, 
              labels = location_types,
              colors = colors, 
              autopct="%1.1f%%", 
              startangle=-90,
             counterclock = True,
             shadow = True)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("../Images/TotalFaresByCityType.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_ride_df = city_ride_df.groupby("type").count()

location_types=["Urban", "Rural", "Suburban"]
fares_list = [total_ride_df.loc[location]["ride_id"] for location in location_types]
colors = ["salmon", "gold", "lightskyblue"]
explode = [0.1,0,0]

# Build Pie Chart
fig = plt.pie(fares_list, 
              explode = explode, 
              labels = location_types,
              colors = colors, 
              autopct="%1.1f%%", 
              startangle=-90,
             counterclock = True,
             shadow = True)
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("../Images/TotalRidesByCityType.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivers_by_city_type_df = city_df.groupby("type").sum()
location_types=["Urban", "Rural", "Suburban"]
driver_count_list = [drivers_by_city_type_df.loc[location]["driver_count"] for location in location_types]
colors = ["salmon", "gold", "lightskyblue"]
explode = [0.1,0,0]

# Build Pie Charts
fig = plt.pie(driver_count_list, 
              explode = explode, 
              labels = location_types,
              colors = colors, 
              autopct="%1.1f%%", 
              startangle=-90,
             counterclock = True,
             shadow = True)
text = plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("../Images/TotalDriversByCityType.png")
```

```python
# Show Figure
plt.show()
```
